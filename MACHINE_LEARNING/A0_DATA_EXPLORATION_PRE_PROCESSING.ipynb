### Problem Overview

1. The dataset includes house prices and features such as bedrooms, bathrooms, flat area, basement area, etc.
2. The goal is to predict the **Sale Price** of a house based on these characteristics.
3. This model can help sellers estimate prices and buyers verify fair value.

---

### Type of Problem

4. This is a **Supervised Learning** problem since the target variable is known.
5. As the target (Sale Price) is continuous, we use **Linear Regression**.
