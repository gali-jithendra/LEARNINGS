#####  Treating outliers of Target variable

```python
plt.scatter(df['ID'],df['Sale Price'])
```

```python
sns.boxplot(df['Sale Price'], vert=True)
```

```python
q1,q2,q3 = df['Sale Price'].quantile([0.25,0.5,0.75])
iqr = q3-q1
ul = q3+1.5*iqr
ll = q1-1.5*iqr
q1
```

```python
def limit_imputer(value):
    if value > ul:
        return ul
    if value < ll:
        return ll
    else:
        return value
```

```python
df['Sale Price'] = df['Sale Price'].apply(limit_imputer)
df[['Sale Price']]
```

```python
df['Sale Price'].describe()
```

```python
sns.boxplot(df['Sale Price'])
```

Only those variables should be selected for building
the model which have high correlation (above a
certain cut-off value), positive or negative, with the
target variable.

If two independent variables that are highly correlated witt
each other, are also correlated with the dependent variabl
then using both independent variables in the model may
result in a suboptimal or poorly understood model.

```python
high_pairs = (corr_df.abs().unstack().sort_values())
high_pairs
```

```python
high_pairs = high_pairs[(high_pairs < 1) & (high_pairs >0.85)]
high_pairs
```

Q) Is there a mathematical or statistical way to compute relationship between a categorical and a numerical variable?   
A) The method of finding out whether a categorical variable is having an impact on the numeric variable - **ANOVA**   
However, ANOVA can not quantify the relationship between
a categorical variable and a numerical variable.

â€¢ ANOVA stands for Analysis of Variance.  
â€¢ It checks if the **means of the target variable across unique values of a categorical variable** are equal or not.  

If p-value < 0.05: There is less than 5% probability
that the difference in means is purely a coincidence.
One can be 95% or more confident that the
difference in means actually exists.
    

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
df = df.rename(columns = {'Sale Price': 'Sale_Price',
                          'Condition of the House' : 'Condition_of_the_House',
                          'Ever Renovated' : 'Ever_Renovated',
                          'Waterfront View' :'Waterfront_View'} )
df.columns
```

```python
mod = ols('Sale_Price ~ Condition_of_the_House', data = df).fit()
Anova_Table = sm.stats.anova_lm(mod, typ = 2)
Anova_Table
```

```python
mod = ols('Sale_Price ~ Ever_Renovated', data = df).fit()
Anova_Table = sm.stats.anova_lm(mod, typ = 2)
Anova_Table
```

```python
mod = ols('Sale_Price ~ Waterfront_View', data = df).fit()
Anova_Table = sm.stats.anova_lm(mod, typ = 2)
Anova_Table
```

```python
mod = ols('Sale_Price ~ Zipcode', data = df).fit()
Anova_Table = sm.stats.anova_lm(mod, typ = 2)
Anova_Table
```

There is a impact of the categorical variable on target variable  
But ANOVA does not tell us how much is the impact

#### Binning and Creation of Dummy Variables  
- Bin and create dummy variables for a
categorical variable.
- The process of transforming a categorical
variable into a set of numerical or boolean variables, called
dummy variables, each of which has values of 0 or 1.  
    
- The number of dummy variables created for a categorical
variable depends on the number of unique values or
levels that categorical variable has.

- If a categorical variable has 'n' levels,
then 'n-1' dummy variables are required.

```python
df = pd.get_dummies(df, columns = ['Condition_of_the_House'], drop_first = True)
```

```python
df[['Condition_of_the_House_Excellent',	'Condition_of_the_House_Fair',	'Condition_of_the_House_Good',
    'Condition_of_the_House_Okay']].head(3)
```

```python
df = pd.get_dummies(df, columns = ['Ever_Renovated','Waterfront_View'], drop_first = True)
```

```python
df[['Ever_Renovated_Yes',	'Waterfront_View_Yes']].head(3)
```

If the number of levels for a categorical variable is more  
(say > 20), then bin these levels into fewer groups before  
creating the dummy variables.  
Zipcode - 71 Levels  
â€¢ 70 dummy variables are required  
Bin zipcodes with similar average Sale Price into groups.  

#### Binning

```python
Zip_Table = df.groupby('Zipcode').agg({'Sale_Price':'mean'}).sort_values('Sale_Price', ascending = True)
Zip_Table.head()
```

```python
Zip_Table['Zipcode'] = pd.cut(Zip_Table['Sale_Price'], bins = 10,
                              labels = ['Zipcode_Group_0',
                                        'Zipcode_Group_1',
                                        'Zipcode_Group_2',
                                        'Zipcode_Group_3',
                                        'Zipcode_Group_4',
                                        'Zipcode_Group_5',
                                        'Zipcode_Group_6',
                                        'Zipcode_Group_7',
                                        'Zipcode_Group_8',
                                        'Zipcode_Group_9'],
                              include_lowest = True)
```

```python
Zip_Table = Zip_Table.drop(columns = 'Sale_Price')
```

```python
Zip_Table.tail()
```

```python
df = pd.merge(df, Zip_Table, left_on = 'Zipcode'
              , how = 'left', right_index = True)
df = df.drop(columns = "Zipcode")
```

```python
df[['ID','Sale_Price', 'Zipcode_y']].head(3)
```

```python
df = pd.get_dummies(df, columns = ['Zipcode_y'], drop_first = True)
```

```python
df[['Zipcode_y_Zipcode_Group_1','Zipcode_y_Zipcode_Group_2','Zipcode_y_Zipcode_Group_3']].head(3)
```

â€¢ Binning can be used both for numerical and categorical variables.
