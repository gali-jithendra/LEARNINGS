```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import scipy.stats as ss
```

###  Create a normally distributed population

```python
np.random.seed(1)
```

```python
a = np.random.normal(10,2,size=10000)
```

```python
sns.kdeplot(a)
```

###  Create a sampling distribution By sampling the normal distribution population

-- Create random samples  
-- Calculate Means   
-- analyze mean distribution.  

```python
b = random.sample(list(a),10) #1 samples
b
```

```python
pd.Series(b).mean()
```

```python
#2nd time running the sample
b = random.sample(list(a),10) #1 samples
b
```

```python
pd.Series(b).mean()
```

-- like wise do it 100 times

```python
def sampling_mean(a,n):
    means = []
    for i in np.arange(0,100,1):
        means.append(np.array(random.sample(list(a),n)).mean())
    return means
```

```python
sampling_mean(a,10)
```

```python
sns.kdeplot(a, color='orange', label='population')
sns.kdeplot(sampling_mean(a,10), color='blue', label='sampling distribution')
plt.legend()
plt.show()
```

###  Create an exponential distribution population

```python
np.random.seed(1)
b = np.random.exponential(size=10000)
sns.kdeplot(b)
```

```python
#Exponential population mean
b.mean()
```

```python
#Sampling mean distribution of exponential population
c = pd.Series(sampling_mean(b,10))
c.mean()
```

```python
sns.kdeplot(b, color='orange', label='exponential population')
sns.kdeplot(sampling_mean(b,10), color='blue', label='sampling distribution')
plt.legend()
plt.show()
```

  -- Central limit Theorem says:  
  Irrespective of the population distribution or whether it is normal or exponential etc the sampling distribution of Sample means from population distribution  
  is always a normal distribution.  


### Hypothesis Testing

#### Right Tail Test

Example there is a skilled development portal for the employees of a company The company has made some changes to the portal to attract the employees  
to spend more time After this change is implemented for a set of small number of employees say 25 employees they have recorded the time  
The employee spends on an average per week That is equal to 62 minutes Now they wanted to understand if they implement a criteria for each employee,  
that they should spend more than 60 minutes on the portal per week Will that succeed Should the project go for launch? 

** Step1: Defining Problem  **

I am hypothesizing to have population mean >=60  
  
The sample data mean M=62  
and its standard deviation = 10  
  
Null hypothesis:  
Ho: M<=60    
Assertion:  
Ha: M>60 #Right tail test

** Step2: Defining Acceptance Criteria **  

Acceptance criteria, alpha = 5% = 0.05

```python
critical_value = ss.t.ppf(0.95,df=(25-1))
critical_value
```

** Step3: Calculate Tstat **  

  As the number of data points are less than 50 we consider to calculate T stat instead of Z stat   
  As the population parameters unknown we use standard error instead of population standard deviation.  
  
tstat =  
                (x-m)   
          -----------------  
             (s/root(n))  

```python
tstat = (62-60)/(10/np.sqrt(25))
tstat
```

INFERENCE  
As Tstat value = 1 is < critical value = 1.71088   
The value of sample mean equal to 62 is falling in the most probable region    
And a not falling in the region of rejection    
 So we cannot reject the null hypothesis    
accept that the average time spending cannot exceed 60 or say average time stays less than or equal to 60 only    
 Hence the project cannot go for a launch   

 2nd Method

```python
ss.t.cdf(x=62,loc=60,scale=10/np.sqrt(25),df=24)
```

The probability of value <=62 occurring on T distribution with scale of 2 and the mean 60 is 83.63%.  
The probability of not occurring <= is (100-83.63) = 16.36%  

As we have said, the significance criteria to only 5% on the extreme right side Here, the value we got is 16.36%.  
 Hence, it is not falling in the extreme right side in the region of rejection We cannot reject the null hypothesis.
