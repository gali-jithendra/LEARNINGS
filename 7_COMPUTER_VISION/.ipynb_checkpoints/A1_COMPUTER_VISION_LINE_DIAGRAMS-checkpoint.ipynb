```python
import numpy as np
import cv2
```

```python
## Create a Black Image
img = np.zeros((512,512,3), np.uint8)
cv2.line(img, (240,330), (511,511), (255,0,0), 5)
cv2.rectangle(img, (240,330), (511,511), (0,255,0), 5)
cv2.circle(img, (240,330), 150, (0,0,255), 5)
pts = np.array([[213,45],[245,35],[370,40],[350,78]],np.int32)
pts = pts.reshape(-1,1,2)
cv2.polylines(img, [pts],True,(0,255,255))

cv2.imshow('sample_image', img)
cv2.waitKey(0)
```

1ï¸âƒ£ np.zeros((512,512,3))  
This creates a NumPy array filled with zeros.  
512, 512 â†’ Height and Width of the image (512 pixels Ã— 512 pixels)  
3 â†’ Number of color channels (B, G, R in OpenCV)  
So this means:  
Create a 512Ã—512 color image.  
  
2ï¸âƒ£ np.uint8  
This defines the data type of each pixel.
uint8 = Unsigned 8-bit integer  
Pixel values range from 0 to 255  
Standard format for images  
  
ðŸŽ¯ What does this line actually create?  
It creates a completely black image of size 512Ã—512.  
Why black?  
Because all pixel values are 0:  
B = 0  
G = 0  
R = 0  
And (0,0,0) = Black in RGB.

--> load an image and display  
  
--> Use the function cv2. imread() to read an image.  
--> The image should be in the working directory or a full path of image should be given.  
  
--> Second argument is a flag which specifies the way image should be read.  
  
--> CV2. IMREAD_COLOR: Loads a color image. Any transparency of image will be neglected. It is the default flag.  
--> CV2.IMREAD_GRAYSCALE: Loads image in grayscale mode  
--> CV2. IMREAD_UNCHANGED: Loads image as such including alpha channel  
--> instead we can pass 1,0, -1 respectively  

### Open_cv

```python
import cv2
img = cv2.imread("Z_COMPUTER_VISION_LINE_DIAGRAMS.jpg", 0)
print(type(img), img.shape, img)
cv2.imshow('image', img)
k = cv2.waitKey(0)
if k == 27:  #wait for ESC Key to exit
    cv2.destroyAllWindows()
elif k == ord('s'):  #wait for 's' key to save and exit
    print("saving the image >>>")
    cv2.imwrite("Z_COMPUTER_VISION_LINE_DIAGRAMS.jpg", img)
    print("Image Saved")
    cv2.destroyAllWindows()
```

### Color_channels

```python
import cv2

image = cv2.imread("Z_COMPUTER_VISION_LINE_DIAGRAMS.jpg")

b = image.copy()
#set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0

g = image.copy()
#set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = image.copy()
#set green and red channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0

#RGB - Blue
cv2.imshow('B-RGB', b)
cv2.waitKey(0)
#RGB - Green
cv2.imshow('B-RGB', g)
cv2.waitKey(0)
#RGB - Red
cv2.imshow('B-RGB', r)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Resize

```python
import cv2
import numpy as np

img = cv2.imread("Z_COMPUTER_VISION_LINE_DIAGRAMS.jpg",0)

res = cv2.resize(img,None,fx=.5, fy=0.5, interpolation = cv2.INTER_CUBIC)

#OR
print(img.shape)
height, width = img.shape[:2]

#res = cv2.resize(img, (300,300), interpolation = cv2.INTER_CUBIC)
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.imshow('sclaed_image', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
