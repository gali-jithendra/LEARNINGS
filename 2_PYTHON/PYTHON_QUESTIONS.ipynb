#### Q20) Multiple Return Values: 
Write a function called calculate_statistics that takes a list of numbers and returns three values: the mean (average),
median (middle value), and mode (most common value) of the list.
Example:  
Input: calculate_statistics([1, 2, 3, 2, 4, 2, 51)  
Output: (2.71, 2, 2)  

```python
def calculate_statistics(_list):
    mean = round(sum(_list)/ len(_list),2)
    _list.sort()
    median = _list[len(_list)//2]
    dictionary = {num: _list.count(num) for num in _list}
    mode_number = max(dictionary.values())
    mode = [key for key, value in dictionary.items() if value == mode_number]
    return mean, median, mode

calculate_statistics([1, 2, 3,3,3, 2, 4, 2, 5])
```

#### Q23) Recursive Function: 
Write a recursive function called factorial that calculates the factorial of a given number. The factorial of n is the product of all
positive integers less than or equal to n (e.g., 5! = 5 × 4 × 3 × 2 × 1 = 120).
Example:  
Input: factorial(5)  
Output: 120  
Input: factorial(0)  
Output: 1  

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

factorial(1)
```

#### Project - Python Calculator

```python
# calculator.py
def add(a, b):
    """Returns the sum of a and b."""
    return a+b

def subtract(a, b):
    """Returns the subtraction of a and b."""
    return a-b

def multiply(a, b):
    """Returns the product of a and b."""
    return a*b

def divide(a, b):
    """Returns the division of a and b; handle the division by zero."""
    if b == 0:
        return "Error: division by zero"
    return a/b

def display_history(history):
    """Display the history of calculations."""
    if not history:
        print("No calculations have been performed yet.")
    else:
        print("\nCalculations history:")
        for record in history:
            op, num1, num2, result = record
            print(f"{num1} {op} {num2} = {result}")


def main():
    # List to save history of calculations
    history = []

    # Menu of options to operations and symbol functions
    operations = {
        "1": ("+", add),
        "2": ("-", subtract),
        "3": ("*", multiply),
        "4": ("/", divide)
    }

    while True:
        print("\n----- Calculator Menu -----")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Multiplication")
        print("4. Division")
        print("5. View Calculation History")
        print("6. Exit")

        choice = input("Enter your choice (1-6): ")

        if choice == "6":
            print("Exiting the calculator. Thank you!")
            break
        elif choice == "5":
            display_history(history)
            continue
        elif choice in operations:
            op_symbol, func = operations[choice]
            num1 = float(input("Enter the first number: "))
            num2 = float(input("Enter the second number: "))
            result = func(num1, num2)
            print("Result: ", result)
            history.append((op_symbol, num1, num2, result))
        else:
            print("Invalid choice. Please select one of the options.")

if __name__ == "__main__":
    main()
```
