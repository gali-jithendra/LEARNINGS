#### List Conversions

```python
my_list = [1,2,3,4]
```

```python
tuple_from_list = tuple(my_list)
tuple_from_list
```

```python
set_from_list = set(my_list)
set_from_list
```

```python
dict_from_list = dict(enumerate(my_list))
dict_from_list
```

```python
for i in enumerate(my_list):
    print(i)                                  
```

```python
dict(((0, 1),))
```

#### Tuple Conversions

```python
my_tuple = (1,2,3,2)
```

```python
list_from_tuple = list(my_tuple)
list_from_tuple
```

```python
set_from_tuple = set(my_tuple)
set_from_tuple
```

```python
dict_from_tuple = dict(enumerate(my_tuple))
dict_from_tuple
```

#### Set Conversions

```python
my_set = {1,2,3,2}
my_set
```

```python
list_from_set = list(my_set)
list_from_set
```

```python
tuple_from_set = tuple(my_set)
tuple_from_set
```

#### Dict Conversions

```python
my_dict = {"a":1,
          "b":2,
          "c":3,
          "d":3}
my_dict
```

```python
list_from_dict = list(my_dict.values())
list_from_dict
```

```python
tuple_from_dict = tuple(my_dict.values())
tuple_from_dict
```

```python
set_from_dict = set(my_dict.values())
set_from_dict
```
