```python
import numpy as np
import cv2
```

```python
## Create a Black Image
img = np.zeros((512,512,3), np.uint8)
cv2.line(img, (240,330), (511,511), (255,0,0), 5)
cv2.rectangle(img, (240,330), (511,511), (0,255,0), 5)
cv2.circle(img, (240,330), 150, (0,0,255), 5)
pts = np.array([[213,45],[245,35],[370,40],[350,78]],np.int32)
pts = pts.reshape(-1,1,2)
cv2.polylines(img, [pts],True,(0,255,255))

cv2.imshow('sample_image', img)
cv2.waitKey(0)
```
