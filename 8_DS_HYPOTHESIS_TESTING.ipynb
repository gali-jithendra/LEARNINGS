#### Right Tail Test

Q) -  There is a skilled development portal of a company, They have made some changes in that portal.  
- And they want each employee must spend at least 60 minutes on that portal per week.  
- For this, they have taken 25 employees sample data And their respective mean is 62 minutes average per week With the standard deviation of 10 minutes.  
- should the project go for launch?  

A)   
Ho: Population.Mean<=60  
Ha: Population.Mean>60   
  
Hypothesized Mean: 60   
Standard Deviation of sample data: 10  
Population parameters are unknown.  
Significance value: 5% = 0.05.  
  
T distribution as data points are <50  
tstat instead of zstat as population parameters are unknown  

```python
critical_value = 
tstat = 
```

```python
critical_value = ss.t.ppf(q=0.95,df=25-1)
critical_value
```

--  1.71 is the value on X axis of standard T distribution which is at the right most region of rejection Alpha.

```python
tstat = (62-60)/(10/np.sqrt(25))
tstat
```

▶️ As 1 is less than 1.7 which means the Tstat is less than critical value, means our hypothesized mean is falling in the most probable region and not in the region of rejection. Hence We cannot reject the null hypothesis.

❌  Project cannot go for the launch.

```python
np.std(Birthweights)
```

```python
ttest_1samp(Birthweights,popmean=3.5)
```

--  Here the P value is very closer to, 0.05 But as this is the two tailed list we have to take it as half


```python
ttest_1samp(Birthweights,popmean=3.5).pvalue/2
```

--  As Alpha is also halved, Alpha = 0.025.  
--  AS 0.0256 > 0.025, its not fallin in the region of rejection.  
❌ Hence we cannot reject the null hypothesis.
