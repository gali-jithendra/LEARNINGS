#### Data normalization
We then normalize the data dimensions so that they are of approximately the same scale.

```python
x_train = x_train / 255
x_test = x_test / 255
```

**Split the data into train/validation/test datasets**

In the earlier step of importing the date, we had 60,000 datasets for training and 10,000 test datasets. Now we further split
train/validation. Here is how each type of dateset is used in deep learning:  
    
 **Training data**  used for training the model  
 **Validation data** - used for tuning the hyperparameters and evaluate the models  
 **Test data**  used to test the model after the model has gone through initial vetting by the validation set.  

```python
x_valid, x_train = x_train[:5000], x_train[:5000]
y_valid, y_train = y_train[:5000], y_train[:5000]
```

#### Feature Scaling

Why validation dataset?

- Predicting the values of Dataset where we have both Independent variables and Target variables allows us to Tune the Model.
- We don't need to worry about past data, beacuse we can check with Validation data (where we have Target Var).
- Also to tune Hyperparameters of teh Model

Why not we do it with Training data only why Validation data needed?

- to check the model accuracy and tune it with data that it is not trained with.

```python
np.random.seed(42)
tf.random.set_seed(42)
```

- 1st covert 2D to 1D 28*28 = 784

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape = [28, 28]))
model.add(keras.layers.Dense(300, activation="relu"))
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation="softmax"))
```

```python
model.summary()
```

Epoch 30:  
accuracy: 0.8934 = 89%  
inaccuracy: 0.11 = 11% 

    - In this 11%, loss: 0.3244 = 32%, we are just 32% away from generating accurate answers

 accuracy   
Accuracy on training data   
Percentage of correct predictions made by the model during training   

    Example:   
    accuracy = 0.8934  89.34% correct on training data   


 val_accuracy   
Accuracy on validation data   
Data the model has not seen during training   
Used to check generalization   

    Example:   
    val_accuracy = 0.8916  89.16% correct on validation data   

 loss   

Predictions  Loss Calculation  Optimizer Updates Weights   

|Metric	  |What it measures                        |
|---------|----------------------------------------|
|Accuracy |% correct predictions                   |
|Loss	  |How far predictions are from true values|

 Average prediction error is 0.3244 based on the chosen loss function   
(e.g., cross-entropy).   

Lower is better.   

Imagine throwing darts:   
Accuracy  Did you hit the board?   
Loss  How far from the center?   

Optimizer tries to move darts toward center.   
Loss = value of the error (cost) function.   

It tells how far the models predictions are from the true values according to a mathematical formula.   
Loss = Error Function Output   


If using categorical cross-entropy:   
Loss =   (true_label  log(predicted_probability))   

If using mean squared error (MSE):   
Loss = average((true  predicted))   

Different problems  different loss functions  different error definitions.   

|Term	|Meaning                                                |
|-------|-------------------------------------------------------|
|Error	|Difference between prediction and true value (concept) |
|Loss	|Mathematical function that quantifies error            |
|Cost	|Average loss over dataset (often used interchangeably) |



Accuracy   
Accuracy = Correct Predictions / Total   
         = 15 / 20   
         = 75%   

 Your understanding here is correct.   

 Loss is NOT 25%   
Loss depends on:   
    Prediction probabilities   
    Confidence of the model   
    Loss function formula   

Not just right vs wrong count.   

     Example to See Why   
    Suppose 5 wrong predictions exist.   
    
    Case 1  Model is slightly wrong   
    Predictions:   
    True: Cat   
    Pred: Cat = 0.45, Dog = 0.55   
    Loss = small   
    
    Case 2  Model is very confident but wrong   
    True: Cat   
    Pred: Cat = 0.01, Dog = 0.99   
    Loss = very large   
    
    Same accuracy  different loss.   

 Important Insight   
Two models can have:   

    Accuracy = 75%   
    Loss = different values   

Because loss measures how wrong + how confident.
