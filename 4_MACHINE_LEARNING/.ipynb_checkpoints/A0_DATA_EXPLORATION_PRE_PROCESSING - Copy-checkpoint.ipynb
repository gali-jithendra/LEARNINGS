#####  Treating outliers of Target variable

```python
plt.scatter(df['ID'],df['Sale Price'])
```

```python
sns.boxplot(df['Sale Price'], vert=True)
```

```python
q1,q2,q3 = df['Sale Price'].quantile([0.25,0.5,0.75])
iqr = q3-q1
ul = q3+1.5*iqr
ll = q1-1.5*iqr
q1
```

```python
def limit_imputer(value):
    if value > ul:
        return ul
    if value < ll:
        return ll
    else:
        return value
```

```python
df['Sale Price'] = df['Sale Price'].apply(limit_imputer)
df[['Sale Price']]
```

```python
df['Sale Price'].describe()
```

```python
sns.boxplot(df['Sale Price'])
```

```python
numerical_columns = ['No of Bathrooms', 'Flat Area (in Sqft)', 
                     'Lot Area (in Sqft)', 'Area of the House from Basement (in Sqft)', 'Latitude',
                     'Longitude', 'Living Area after Renovation (in Sqft)']
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'median')
df[numerical_columns] = imputer.fit_transform(df[numerical_columns])
```

```python
df.info()
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df['Zipcode'] = imputer.fit_transform(df[['Zipcode']])
```

```python
df.info()
```

```python
df['Zipcode'] = df['Zipcode'].astype(object)
```

```python
df['No of Times Visited'].unique()
```

```python
mapping = {'nan':0,
           'Thrice':'3',
           'Four':'4',
           'Twice':'2',
           'Once':'1'}
df['No of Times Visited'] = df['No of Times Visited'].map(mapping)
```

-  Instead of Renevated Year, lets have data how long the house was renovated.

```python
df['Ever Renovated'] = np.where(df['Renovated Year']==0, 'No', 'Yes')
```

```python
df['Purchase Year'] = pd.DatetimeIndex(df['Date House was Sold']).year
df['Years Since Renovation'] = np.where(df['Ever Renovated']=='Yes',
                                        abs(df['Purchase Year']-df['Renovated Year']),0)
```

```python
# Renovated Year, Purchase Year - âŒ 

df.drop(columns=['Purchase Year','Renovated Year','Date House was Sold'], inplace=True)
```

```python
transformed_df = df[["ID","Sale Price","No of Bedrooms","No of Bathrooms",	"Flat Area (in Sqft)","Lot Area (in Sqft)","No of Floors","Overall Grade","Area of the House from Basement (in Sqft)","Basement Area (in Sqft)","Age of House (in Years)",	"Latitude","Longitude","Living Area after Renovation (in Sqft)","Lot Area after Renovation (in Sqft)","Years Since Renovation"]]
transformed_df
```

```python
corr_df = transformed_df.drop(columns = ['ID']).corr()
corr_df
```

Only those variables should be selected for building
the model which have high correlation (above a
certain cut-off value), positive or negative, with the
target variable.

If two independent variables that are highly correlated witt
each other, are also correlated with the dependent variabl
then using both independent variables in the model may
result in a suboptimal or poorly understood model.

```python
high_pairs = (corr_df.abs().unstack().sort_values())
high_pairs
```

```python
high_pairs = high_pairs[(high_pairs < 1) & (high_pairs >0.85)]
high_pairs
```
